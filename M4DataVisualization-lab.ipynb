## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

df1 = pd.read_sql("SELECT * FROM master", conn)
df1

df1['ConvertedComp'].plot.hist(bins=30, alpha=0.5)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df1['Age'].plot.box()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.scatter(df1['Age'],df1['WorkWeekHrs'])
plt.title('Scatter plot x=Age and Y=Workweek hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(df1['WorkWeekHrs'],df1['CodeRevHrs'],s=df1['Age'],alpha=0.5, c='red')
plt.title('Bubble plot of Workweek hours and CodeRevHrs with Age as buble size')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
df2 = pd.read_sql("SELECT * FROM DatabaseDesireNextYear", conn)
df2
df2_reset=df2['DatabaseDesireNextYear'].value_counts().reset_index(name='Counts')
df2_top5 = df2_reset[:5]
df2_top5
plt.pie(df2_top5['Counts'],labels=df2_top5['index'],autopct='%.2f')
plt.title('Top 5 databases respondents wish to learn next year')
plt.show()
```

```python
dfQ = pd.read_sql("SELECT * FROM LanguageDesireNextYear", conn)
dfQ
dfQ_reset=dfQ['LanguageDesireNextYear'].value_counts().reset_index(name='Counts')
dfQ_top5 = dfQ_reset[:5]
dfQ_top5
```

```python
dfQ2 = pd.read_sql("SELECT * FROM DatabaseWorkedWith", conn)
dfQ2
dfQ2_count=dfQ2['DatabaseWorkedWith'].value_counts()
dfQ2_count
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.




```python
# your code goes here

QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df3 = pd.read_sql_query(QUERY,conn)
df3_groupped = df3.groupby('Age').median()
df3_groupped.plot(kind='bar',stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age,ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df4 = pd.read_sql_query(QUERY,conn)
df4_groupped=df4.groupby('Age').median()
df4_groupped.plot(kind='line')
plt.title('Median ConvertedComp for 45-60 age groups')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch
FROM master
"""
df5 = pd.read_sql_query(QUERY,conn)
df5_reset=df5['MainBranch'].value_counts().rename_axis('MainBranch').reset_index(name='Counts')
df5_new = df5_reset.set_index('MainBranch')
df5_new
df5_new.plot(kind='barh')
plt.title('MainBranch')
plt.show()
```
